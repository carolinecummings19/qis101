```python
# Cell 1
# """rlc_circuit.py"""

# Task 18-02
# Caroline Cummings

from __future__ import annotations

import typing

import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint  # type: ignore

if typing.TYPE_CHECKING:
    from typing import Any
    from matplotlib.axes import Axes

%matplotlib widget

def model (I, t) -> Any: #type: ignore
    """Function that returns I'(t) and I"(t)"""
    # R = 0.1 Ohms, L = 0.01 Henries, C = 0.01 Farads
    return [I[1], -10_000 * I[0] - 10 * I[1]] #type: ignore


def plot(ax: Axes) -> None:
    # Using odeint to find and plot current based on 2nd order ODE
    # https://cmps-people.ok.ubc.ca/jbobowsk/Python/html/Jupyter%20Second%20Order%20ODEs.html
    
    time = np.arange(0, 1, 0.001)

    # Pass in model function, initial conditions, and time array
    current, deriv = odeint(model, [1, 0], time).T #type: ignore
    
    # Plot and label
    ax.axhline(y=0.0, color="lightgray")
    ax.plot(time, current) #type: ignore
    ax.set_title('Resistor-Inductor-Capacitor Circuit')
    ax.set_xlabel("Time (secs)")
    ax.set_ylabel("Current (Amps)")

def main() -> None:
    plt.figure(" ")
    plot(plt.axes())
    plt.show()


main()
```
