## Part 1
Scatter plot using equal aspect ratio of the complex function: <br>
&emsp;&emsp;$f(z) = z^2 + iz + 1$, $|Re(z)| \leq 4$, $0 \leq Im(z) \leq 2$

```python
from __future__ import annotations

import typing
import matplotlib.pyplot as plt
import numpy as np

if typing.TYPE_CHECKING:
    from matplotlib.axes import Axes
    from numpy.typing import NDArray

%matplotlib widget 

def complex_function (z: complex) -> complex:
    return z**2 + 1j * z + 1

def plot(ax: Axes) -> None:
    """Plots scatter plot of complex function"""
    
    # List of complex numbers - real and imaginary parts are bounded
    # From D. Biersach (originally had range(-4, 4) and range(0, 2))
    real_bounds: NDArray[np.float_] = np.linspace(-4, 4, 50) 
    imag_bounds: NDArray[np.float_] = np.linspace(0, 2, 50) 
    complex_numbers: list[complex] = [complex(a, b) for a in real_bounds \
                                      for b in imag_bounds]

    # Evaluate the function for each complex number
    results: list[complex] = [complex_function(z) for z in complex_numbers]

    # Separate the real and imaginary parts of the results
    x: list[float] = [z.real for z in results]
    y: list[float] = [z.imag for z in results]
    
    # Plot and label
    ax.scatter(x, y, s=1) #make points smaller
    ax.set_title(r"Scatter Plot of $f(z) = z^2 + iz + 1$")
    ax.set_ylabel('Imaginary')
    ax.set_xlabel('Real')

    # equal aspect ratio: 
    ax.axis('equal')


def main() -> None:
    plot(plt.axes())
    plt.show()

main()
```

## Part 2
Calculate and display how many Gaussian integers z there are where: <br>
• $|Re(f(z))| \leq 10$, $|Im(f(z))| \leq 2$

```python
def gaussian_ints () -> None:
    count: int = 0

    # From D. Biersach
    for x in range (-50, 50):
        for y in range (-50, 50):
            if abs(x**2 - y**2 - y + 1) <= 10 and abs(2 * x * y + x) <= 10:
                count += 1
    
    print(f"There are {count} Gaussian integers in the given conditions.")

gaussian_ints()
```
